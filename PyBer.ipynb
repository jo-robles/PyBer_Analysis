```python
#Combining the data into a single dataset.

pyber_data_df = pd.merge(ride_data_df, city_data_df, how='left', on=['city','city'])

#Display the DataFrame

pyber_data_df.head()
```

```python
#Creating an Urban city DataFrame.

urban_cities_df = pyber_data_df[pyber_data_df['type'] == 'Urban']
urban_cities_df.head()
```

```python
#Creating a Suburban and rural city DataFrame.

suburban_cities_df = pyber_data_df[pyber_data_df['type'] == 'Suburban']

rural_cities_df = pyber_data_df[pyber_data_df['type'] == 'Rural']

rural_cities_df
```

```python
#Getting the number of rides for urban cities.

urban_ride_count = urban_cities_df.groupby(['city']).count()['ride_id']
urban_ride_count.head()
```

```python
#Getting the number of rides for the rural and suburban cities.

suburban_ride_count = suburban_cities_df.groupby(['city']).count()['ride_id']
rural_ride_count = rural_cities_df.groupby(['city']).count()['ride_id']
```

```python
#Getting the average of the fares for the urban cities.

urban_avg_fare = urban_cities_df.groupby(['city']).mean()['fare']
urban_avg_fare.head()
```

```python
#Getting the average of the fares for the rural and suburban cities.

suburban_avg_fare = suburban_cities_df.groupby(['city']).mean()['fare']

rural_avg_fare = rural_cities_df.groupby(['city']).mean()['fare']
```

```python
#Getting the average number of drivers for the urban cities.

urban_driver_count = urban_cities_df.groupby(['city']).mean()['driver_count']
urban_driver_count
```

```python
#Getting the average number of drivers for the rural and suburban cities.

suburban_driver_count = suburban_cities_df.groupby(['city']).mean()['driver_count']

rural_driver_count = rural_cities_df.groupby(['city']).mean()['driver_count']
```

```python
#Building the scatter plots for urban cities.

plt.scatter(urban_ride_count, urban_avg_fare, s=10*urban_driver_count, c = 'coral', edgecolor = 'black',linewidth = 1, alpha = .8, label = 'Urban')
plt.title('Pyber Ride Sharing Data (2019)')
plt.ylabel('Average Fare ($)')
plt.xlabel('Total Number of Rides (Per City)')
plt.grid(True)
#Adding the legend.
plt.legend()
```

```python
#Building the scatter plots for suburban cities.

plt.scatter(suburban_ride_count, suburban_avg_fare, s=10*suburban_driver_count, c = 'skyblue', edgecolor = 'black',linewidth = 1, alpha = .8, label = 'Suburban')
plt.title('Pyber Ride Sharing Data (2019)')
plt.ylabel('Average Fare ($)')
plt.xlabel('Total Number of Rides (Per City)')
plt.grid(True)
#Adding the legend.
plt.legend()
```

```python
#Building the scatter plots for rural cities.

plt.scatter(rural_ride_count, rural_avg_fare, s=10*rural_driver_count, c = 'gold', edgecolor = 'black',linewidth = 1, alpha = .8, label = 'Rural')
plt.title('Pyber Ride Sharing Data (2019)')
plt.ylabel('Average Fare ($)')
plt.xlabel('Total Number of Rides (Per City)')
plt.grid(True)
#Adding the legend.
plt.legend()
```

```python
#Combining the scatter charts for each city type:

plt.scatter(urban_ride_count, urban_avg_fare, s=10*urban_driver_count, c = 'coral', edgecolor = 'black',linewidth = 1, alpha = .8, label = 'Urban')
plt.scatter(suburban_ride_count, suburban_avg_fare, s=10*suburban_driver_count, c = 'skyblue', edgecolor = 'black',linewidth = 1, alpha = .8, label = 'Suburban')
plt.scatter(rural_ride_count, rural_avg_fare, s=10*rural_driver_count, c = 'gold', edgecolor = 'black',linewidth = 1, alpha = .8, label = 'Rural')

#Formatting the graph
plt.title('Pyber Ride Sharing Data (2019)', fontsize = 20)
plt.ylabel('Average Fare ($)',fontsize = 12)
plt.xlabel('Total Number of Rides (Per City)', fontsize = 12)
plt.grid(True)

#Creating a legend.
lgnd = plt.legend(fontsize = '12', mode = 'Expanded', scatterpoints = 1, loc = 'best', title = 'City Types')
lgnd.legendHandles[0]._sizes = [75]
lgnd.legendHandles[1]._sizes = [75]
lgnd.legendHandles[2]._sizes = [75]
lgnd.get_title().set_fontsize(12)

#Incorporate a text label about circle_size.
plt.text(42,35,'Note: Circle size coorelates with driver count per city.', fontsize='12')

#Save the figure.
plt.savefig('Analysis/Fig1.png')

#showing the plot
plt.show()
```

```python
#Skill drill - Asking to calculate mean, median, mode for urban, suburban, rural driver count. 

#Getting the driver count data for suburban and rural (urban already made)
suburban_drivers = suburban_cities_df['driver_count']
rural_drivers = rural_cities_df['driver_count']

#Getting the averages
average_urban_drivers = np.mean(urban_drivers)
average_suburban_drivers = np.mean(suburban_drivers)
average_rural_drivers = np.mean(rural_drivers)

#Getting the medians
median_urban_drivers = np.median(urban_drivers)
median_suburban_drivers = np.median(suburban_drivers)
median_rural_drivers = np.median(rural_drivers)

#Getting the modes
mode_urban_drivers = sts.mode(urban_drivers)
mode_suburban_drivers = sts.mode(suburban_drivers)
mode_rural_drivers = sts.mode(rural_drivers)

#Printing out averages:
print(f'The averages for urban, suburban and rural drivers are as follows: Urban - {average_urban_drivers:.0f}, Suburban - {average_suburban_drivers:.0f}, Rural - {average_rural_drivers:.0f}')

#Printing out the medians:
print(f'The medians for urban, suburban and rural drivers are as follows: Urban - {median_urban_drivers:.0f}, Suburban - {median_suburban_drivers:.0f}, Rural - {median_rural_drivers:.0f}')

#Printing out the modes:
print(f'The mode for urban, suburban and rural drivers are as follows: Urban - {mode_urban_drivers}, Suburban - {mode_suburban_drivers}, Rural - {mode_rural_drivers}')
```

```python
#Next step, getting the sum of the fares for each city type.

sum_fares_by_type = pyber_data_df.groupby(['type']).sum()['fare']
sum_fares_by_type
```

```python
#Getting the sum of all the fares.

total_fares = pyber_data_df['fare'].sum()
total_fares
```

```python
#Calculating the percentage of fare for each city type.

type_percents = 100 * sum_fares_by_type / total_fares
type_percents
```

```python
#Another way of doing this in one line.

type_percents = 100 * pyber_data_df.groupby(['type']).sum()['fare'] / pyber_data_df['fare'].sum()

type_percents
```

```python
#Building the percentages of fares by city type pie chart.

plt.pie(type_percents, labels = ['Rural','Suburban','Urban'])
plt.show()
```

```python
#Adding details to the pie chart.

plt.pie(type_percents,
       labels = ['Rural','Suburban','Urban'],
       colors = ['gold','lightskyblue','lightcoral'],
       explode = [0,0,0.1],
       autopct = '%1.1f%%',
       shadow = True,
       startangle = 150)
plt.title('% of Total Fares by City Type')

#Showing figure
plt.show()
```

```python
#To change some default parameters, we need to import that matplotlib library

#Import mpl to change the plot configurations using rcParams.
import matplotlib as mpl

#Building the Pie Chart
plt.subplots(figsize=(10,6))

plt.pie(type_percents,
       labels = ['Rural','Suburban','Urban'],
       colors = ['gold','lightskyblue','lightcoral'],
       explode = [0,0,0.1],
       autopct = '%1.1f%%',
       shadow = True,
       startangle = 150)
plt.title('% of Total Fares by City Type')

#Changing the default font size from 10 to 14.
mpl.rcParams['font.size'] = 14

#Saving the figure
plt.savefig('Analysis/Fig5.png')

#Showing figure
plt.show()
```
