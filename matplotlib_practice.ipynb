```python
%matplotlib inline
```

```python
#Import dependencies

import matplotlib.pyplot as plt
```

```python
#Setting the x-axis to a list of strings for each month.

x_axis = ['Jan','Feb','Mar','April','May','June','July','Aug','Sept','Oct','Nov','Dec']

#Setting the y-axis to a list of floats as the total fare in US dollars accumulated for each month.
y_axis = [10.02, 23.24, 39.20, 35.42, 32.34, 27.04, 43.82, 10.56, 11.85, 27.90, 20.71, 20.09]
```

```python
#Creating the plot.

plt.plot(x_axis, y_axis)
```

```python
#Creating the plot utilizing ax.plt()
fig, ax = plt.subplots()
ax.plot(x_axis,y_axis)
```

```python
#Creating the plot with ax.plt() as another method.
fig = plt.figure()
ax = fig.add_subplot()
ax.plot(x_axis, y_axis)
```

```python
#Creating the plot.
plt.plot(x_axis,y_axis)
plt.show()
```

```python
#Creating the plot and adding a label for the legend.
plt.plot(x_axis, y_axis, marker = '*', color = 'blue', linewidth = 2, label='Boston')

#Creating labels for the x and y axes.
plt.xlabel('Date')
plt.ylabel('Fare($)')

#Setting the y-limit between 0 and 45.
plt.ylim(0,45)

#Creating a title.
plt.title('PyBer Fare by Month')

#Adding a grid.
plt.grid()

#Adding the legend.
plt.legend()
```

```python
#Using object oriented methods via the skill drill.

fig = plt.figure()
ax = fig.add_subplot()
ax.plot(x_axis, y_axis, color = 'green', marker = 'd', linewidth = 2, label = 'Boston')
ax.grid()
ax.set_xlabel('Date')
ax.set_ylabel('Fares ($)')
ax.legend()
ax.set_ylim(0,45)
```

```python
#Creating the plot.

plt.bar(x_axis, y_axis)
```

```python
#Formatting the bar-chart

plt.bar(x_axis,y_axis, color = 'green', label = 'Boston')

# Creating labels

plt.xlabel('Date')
plt.ylabel('Fare($)')

#Creating a title.
plt.title('PyBer Fare by Month')

#Adding the legend.
plt.legend()
```

```python
#Creating a horizontal bar chart.

plt.barh(x_axis,y_axis)
```

```python
# Demonstrating how to switch x and y

plt.barh(y_axis, x_axis)
```

```python
#Going back to previous bar chart and fixing the order of the Y-Axis

plt.barh(x_axis, y_axis)
plt.gca().invert_yaxis()
```

```python
#Formatting the bar-chart (horizontal), skill drill

plt.barh(x_axis,y_axis, color = 'magenta', label = 'Boston')
plt.gca().invert_yaxis()

# Creating labels

plt.xlabel('Date')
plt.ylabel('Fare($)')

#Creating a title.
plt.title('PyBer Fare by Month')

#Adding the legend.
plt.legend()
```

```python
#Demonstrating how to create a bar chart via object-oriented interface method.

#Creating a plot with the ax.plt()

fig, ax = plt.subplots()
ax.bar(x_axis,y_axis)
```

```python
#showing how to put the data on opposite axes.

fig, ax = plt.subplots()
ax.barh(y_axis, x_axis)
```

```python
#Skill Drill. How to create a horizontal bar chart.

fig, ax = plt.subplots()

#Creating the bar chart, setting the color to cyan and adding the label - Chicago
ax.barh(x_axis,y_axis, color = 'cyan', label = 'Chicago')

#Creating the X-Label
ax.set_ylabel('Date')

#Creating the Y-Label
ax.set_xlabel('Fares ($)')

#Inverting the y
ax.invert_yaxis()

#Adding the legend
ax.legend()
```

```python
#Creating a scatterplot from the data.

plt.plot(x_axis,y_axis, 'o')
```

```python
#Another method to create a scatterplot.

plt.scatter(x_axis,y_axis)
```

```python
#Skill drill

#Setting up the scatter plot, changing the color to red and adding the Chicago label.
plt.scatter(y_axis,x_axis,color='red',label = "Chicago")

#Inverting the y-axis
plt.gca().invert_yaxis()

#Adding the y-axis label
plt.ylabel('Date')

#Adding the x-axis label
plt.xlabel('Fare($)')

#Setting the x-axis limiter
plt.xlim(0,50)

#Showing the legend
plt.legend()
```

```python
#Creating a bubble chart

plt.scatter(x_axis,y_axis, s=y_axis)
```

```python
#Switching up the items.

y_axis_larger=[]

for data in y_axis:
    y_axis_larger.append(data*3)

plt.scatter(x_axis,y_axis,s=y_axis_larger)
```

```python
#Doing the above without using a for looop.

plt.scatter(x_axis,y_axis, s = [i * 3 for i in y_axis])
```

```python
#Creating a scatter chart utilizing object-oriented interface.

fig, ax = plt.subplots()
ax.scatter(x_axis,y_axis)
```

```python
#Creating a bubble chart utilizing object-oriented interface.

fix, ax = plt.subplots()
ax.scatter(x_axis, y_axis, s = y_axis)
```

```python
#Skill drill for bubblechart.

fig, ax = plt.subplots()
ax.scatter(y_axis, x_axis, color = 'skyblue', s = [i * 5 for i in y_axis], linewidth = 2, label = 'Boston', alpha = .2, edgecolors = 'black')
ax.invert_yaxis()
ax.legend()
ax.set_xlim(0,50)
ax.set_title('PyBer Fare by Month')
ax.set_ylabel('Date')
ax.set_xlabel('Fare ($)')
```

```python
#Creating a pie chart.
plt.pie(y_axis, labels = x_axis)
plt.show()
```

```python
explode_values = (0, 0, 0, 0, 0, 0, 0.2, 0, 0, 0, 0, 0)
plt.subplots(figsize=(8,8))
plt.pie(y_axis, explode = explode_values, labels = x_axis, autopct = '%.1f%%')
plt.show()
```

```python
#Assigning 12 colors, one for each month.
colors = ["slateblue", "magenta", "lightblue", "green", "yellowgreen", "greenyellow", "yellow", "orange", "gold", "indianred", "tomato", "mistyrose"]

#Creating the explode value
explode_values = (0, 0, 0, 0, 0, 0, 0.2, 0, 0, 0, 0, 0)

#Increasing the size of the figure.
plt.subplots(figsize=(8,8))

#Creating the pie-chart with explode, colors, labels and percentage formats.
plt.pie(y_axis, explode = explode_values, colors = colors, labels = x_axis, autopct = '%.1f%%')

#Displaying the pie chart
plt.show()
```

```python
#Doing the same pie chart with object oriented interface.

fig, ax = plt.subplots()
ax.pie(y_axis, labels = x_axis)
plt.show()
```

```python
#Skill Drill
fig, ax = plt.subplots(figsize = (8,8))
explode_values = (0, 0, .2, 0, 0, 0, 0.2, 0, 0, 0, 0, 0)
colors = ["slateblue", "magenta", "lightblue", "green", "yellowgreen", "greenyellow", "yellow", "orange", "gold", "indianred", "tomato", "mistyrose"]
ax.pie(y_axis, labels = x_axis, startangle = -90, explode = explode_values, colors = colors, shadow = True, autopct = '%.1f%%')
ax.invert_yaxis()
plt.show()
```
